```python
import matplotlib.pyplot as plt
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.bar(years, runs)
plt.xlabel("years")
plt.ylabel("Runs")
plt.title("Basic Bar chart")
plt.show()
```

```python
##horizontal bar

players = ["Kohli", "Rohit", "Dhoni", "Gill"]
runs = [12000, 11000, 10000, 2500]

plt.barh(players, runs)

plt.xlabel("Players")
plt.ylabel("Runs")
plt.title("Runs Comparison")

plt.show()
```

```python
import numpy as np
```

```python
subjects = ["Maths", "Science", "English"]

rahul = [85, 90, 80]
priya = [88, 85, 92]
```

```python
x= np.arange(len(subjects))
```

```python
width=0.35
```

```python
plt.bar(x-width, rahul, width=width, label="Rahul")
plt.bar(x+width, priya, width=width, label="Priya")
plt.xticks(x, subjects)
plt.legend()
plt.tight_layout()
plt.show()
```
