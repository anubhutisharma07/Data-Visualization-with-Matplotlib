```python
import matplotlib.pyplot as plt
```

```python
study_hours = [1, 2, 3, 4, 5, 6, 7, 8, 9]
exam_scores = [40, 45, 50, 55, 60, 65, 75, 85, 90]
```

```python
plt.scatter(study_hours, exam_scores)
plt.xlabel("Study hours")
plt.ylabel("Exam score")
plt.title("Student Records")
plt.grid(True)

plt.show()
```

```python
stu_h = [2, 3, 4, 5, 6, 7]
exam_s = [45, 50, 55, 65, 75, 85]
```

```python
plt.scatter(stu_h, exam_s, color="orange", s=250)
plt.xlabel("Hours")
plt.ylabel("Score")
plt.title("Study Hours vs Scores")
plt.grid(True)
plt.show()
```
