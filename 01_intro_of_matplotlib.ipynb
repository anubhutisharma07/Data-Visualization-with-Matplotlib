```python
                                                    #Adding X and Y Labels
```

```python
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010]
runs = [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500]
```

```python
plt.plot(years, runs)
```

```python
plt.plot(years, runs)
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Sachin Tendulkar's Yearly Runs")
plt.show()
```

```python
                                              #Multiple Lines in One Plot
```

```python
kohli = [0, 0, 500, 800, 1100, 1300, 1500, 1800, 1900, 2100]
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
```

```python
plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")
```

```python
plt.plot(years, kohli, label="Virat Kohli")
plt.plot(years, sehwag, label="Virender Sehwag")
plt.xlabel("Year")
plt.ylabel("Runs Scored")
plt.title("Performance Comparison")
plt.legend()
plt.show()
```

```python
student= ["A", "B", "C", "D"]
marks= [80, 65, 90, 75]
```

```python
months = ["Jan", "Feb", "Mar", "Apr"]

product_A = [100, 120, 110, 140]
product_B = [90, 130, 105, 150]
```

```python
plt.plot(months, product_A, color="red", marker="*", label="product_A")
plt.plot(months, product_B, color="Green", linestyle="--", label="product_B")
plt.legend()
plt.grid(True)
plt.show()
```

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri"]

delhi = [38, 40, 39, 42, 41]
mumbai = [32, 33, 31, 34, 35]
jaipur = [36, 38, 37, 40, 39]
```

```python
plt.plot(days, delhi, label="Delhi")
plt.plot(days, mumbai, label="Mumbai")
plt.plot(days, jaipur, label="jaipur")
plt.xlabel("Days")
plt.ylabel("Temp")
plt.title("Temperature Comparison")
plt.grid(True, color='grey', linewidth=1.4, linestyle='-.')
plt.legend()
plt.show()
```
